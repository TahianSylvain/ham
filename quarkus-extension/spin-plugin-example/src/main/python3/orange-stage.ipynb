*Gain the data_sets and test_data from kaggle

```python
import numpy as np
import pandas as pd

datasets = "./drive/MyDrive/Colab Notebooks/Cake_Datasets_Train_Labels.csv"
df_data = pd.read_csv(datasets)
df_data
```

```python

# prompt: let's extract ingredients for training, and nb_body, quality, cost for label

df_X = df_data[['All-purpose flour', 'Granulated Sugar','Cocoa powder', "Baking Soda",
                "Baking powder", "Eggs","Vegetable oil", "Milk","Vanilla extract",
                "Chocolat Chips","Salt", "Unsalted Butter", 'Lemon', 'Carrot', 'Strawberry']
               ].to_numpy()
df_Y = df_data[['Nb_Body', 'Quality Rate', 'Cost']].to_numpy()


"""-----------------------------------------------------------------------------------------------------"""
# prompt: and import the lib that shows 3D plot to render the weighs and biases
# !pip install holoviews

import holoviews as hv
from holoviews import opts

hv.extension('plotly'); print(hv.__version__)

scatter3d = hv.Path3D(df_Y, kdims=['Nb_Body', 'Quality Rate', 'Cost'])

# Display the plot
scatter3d.opts(color='orangered', colorbar=True, height=600, width=600)
```

```python

# Adding gradient descent of Nerkov
model = tf.keras.Sequential([
    # tf.keras.layers.Dense(64, activation='relu', input_shape=(15,)),

    tf.keras.Input(shape=(15,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(3) # regression loss func [mse or mae]
])
model.add(tf.keras.layers.Flatten())
model.summary()
```

Creating a rule based learning algorithm

```python
model.compile(loss='mse', optimizer='adam', metrics=['mae'])
"""model.compile( loss='sparse_categorical_crossentropy', metrics=['accuracy'],
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=True),)
"""


# Train the model
model.fit(df_X, df_Y, epochs=10, shuffle=True)
```
